```python
# NumPyを使った基本統計計算

import numpy as np
import pandas as pd

# Titanicデータの読み込み
df = pd.read_csv("titanic3.csv")

# 欠損値を埋める(中央値で埋める例)
df['age'] = df['age'].fillna(df['age'].median())

# 'age'列をNumPy配列に変換
age_data = df['age'].values

# NumPyを使って統計値を計算
mean_age = np.mean(age_data)  # 平均
median_age = np.median(age_data)  # 中央値
std_age = np.std(age_data)  # 標準偏差
var_age = np.var(age_data)  # 分散

# 結果を表示
print("平均年齢:", mean_age)
print("中央値年齢:", median_age)
print("標準偏差年齢:", std_age)
print("分散年齢:", var_age)
```

```python
# NumPyで年齢分布の可視化

import matplotlib.pyplot as plt

# ヒストグラムで分析を可視化
plt.hist(age_data, bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution (Titanic Dataset)')
plt.show
```

```python
from sklearn.datasets import load_iris
import pandas as pd

# Irisデータセットを読み込む
iris = load_iris()

# DataFrameに変換
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target

# データ概要の表示
print(df_iris.head())
print(df_iris.describe())
```

```python
# 特徴量ごとの分布を確認
import seaborn as sns

# 特徴量ごとにヒストグラムを作成
df_iris_melt = pd.melt(df_iris, id_vars='target', var_name='features', value_name='value')

plt.figure(figsize=(12,6))
sns.histplot(data=df_iris_melt, x='value', hue='features', kde=True, element='step')
plt.title('Feature Distribution (Iris Dataset)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 散布図で特徴量間の関係を可視化
# sepal Lengthとsepal widthの関係を散布図で表示
sns.scatterplot(x=df_iris['sepal length (cm)'], y=df_iris['sepal width (cm)'], hue=df_iris['target'], palette='viridis')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Target', labels=iris.target_names)
plt.show()
```
